```python
import seaborn as sns
import matplotlib.pyplot as plt

# Check for missing values
missing_values = df.isnull().sum()
print("Missing values:\n", missing_values)

# Visualize the distribution of the 'Survived' column
sns.countplot(x='Survived', data=df)
plt.title('Survival Count')
plt.show()

# Visualize the distribution of 'Pclass'
sns.countplot(x='Pclass', data=df)
plt.title('Passenger Class Distribution')
plt.show()

# Visualize the distribution of 'Sex'
sns.countplot(x='Sex', data=df)
plt.title('Gender Distribution')
plt.show()

# Visualize survival rate by 'Pclass'
sns.barplot(x='Pclass', y='Survived', data=df)
plt.title('Survival Rate by Passenger Class')
plt.show()

# Visualize survival rate by 'Sex'
sns.barplot(x='Sex', y='Survived', data=df)
plt.title('Survival Rate by Gender')
plt.show()

# Visualize survival rate by 'Age'
plt.figure(figsize=(10, 6))
sns.histplot(df[df['Survived'] == 1]['Age'], bins=20, kde=False, color='green', label='Survived')
sns.histplot(df[df['Survived'] == 0]['Age'], bins=20, kde=False, color='red', label='Did not survive')
plt.legend()
plt.title('Survival Rate by Age')
plt.show()

# Visualize survival rate by 'SibSp' (number of siblings/spouses aboard)
sns.barplot(x='SibSp', y='Survived', data=df)
plt.title('Survival Rate by Number of Siblings/Spouses Aboard')
plt.show()

# Visualize survival rate by 'Parch' (number of parents/children aboard)
sns.barplot(x='Parch', y='Survived', data=df)
plt.title('Survival Rate by Number of Parents/Children Aboard')
plt.show()

# Visualize survival rate by 'Embarked'
sns.barplot(x='Embarked', y='Survived', data=df)
plt.title('Survival Rate by Embarked Port')
plt.show()

# Visualize fare distribution
plt.figure(figsize=(10, 6))
sns.histplot(df['Fare'], bins=30, kde=True)
plt.title('Fare Distribution')
plt.show()

# Visualize correlation matrix
plt.figure(figsize=(12, 8))
sns.heatmap(df.corr(), annot=True, cmap='coolwarm', fmt='.2f')
plt.title('Correlation Matrix')
plt.show()
```
